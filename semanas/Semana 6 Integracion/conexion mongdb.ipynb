```python
from pymongo import MongoClient
from pymongo.errors import ServerSelectionTimeoutError
import os

# Detectar automáticamente si estamos en Docker o local
def get_mongo_client():
    # Primero intentamos con 'mongodb' (para Docker)
    try:
        client = MongoClient('mongodb://mongodb:27017/', serverSelectionTimeoutMS=2000)
        client.admin.command('ping')
        print("✅ Conectado a MongoDB en DOCKER (mongodb:27017)")
        return client
    except:
        pass
    
    # Si falla, intentamos con localhost
    try:
        client = MongoClient('mongodb://localhost:27017/', serverSelectionTimeoutMS=2000)
        client.admin.command('ping')
        print("✅ Conectado a MongoDB en LOCAL (localhost:27017)")
        return client
    except:
        pass
    
    # Si todo falla
    print("❌ No se pudo conectar a MongoDB")
    print("\n📌 Posibles soluciones:")
    print("   1. Ejecuta: docker-compose up -d mongodb")
    print("   2. O instala MongoDB local: sudo apt install mongodb")
    print("   3. O cambia a una base de datos local (SQLite)")
    return None

# Usar la función
client = get_mongo_client()

if client:
    db = client['TiendaBigData']
    coleccion = db['despegardb']
    print("📊 Colección 'despegardb' lista para usar")
    
    # Prueba de inserción
    test_doc = {"test": True, "timestamp": "2024-01-01"}
    coleccion.insert_one(test_doc)
    print("✅ Documento de prueba insertado correctamente")
else:
    print("⚠️ No hay conexión a MongoDB, los datos NO se guardarán")
```
