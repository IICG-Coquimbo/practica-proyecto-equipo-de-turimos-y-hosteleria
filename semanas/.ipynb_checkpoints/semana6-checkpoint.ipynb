```python
# --- PROYECTO BIG DATA: SEMANA 6 ---
# Estudiante: Angelo Rojo
# Objetivo: Procesamiento y limpieza de datos masivos con Apache Spark

from pyspark.sql import SparkSession
from pyspark.sql.functions import col, upper, trim, round, mean
import certifi

# 1. CONFIGURACIÓN DE LA SESIÓN DE SPARK
# Se incluye el conector oficial de MongoDB para que Spark pueda leer desde la nube.
# Nota: La primera vez que se ejecute, Spark descargará los paquetes necesarios.
spark = SparkSession.builder \
    .appName("Limpieza_AngeloRojo_Semana6") \
    .config("spark.mongodb.read.connection.uri", "mongodb+srv://AngeloRojo:azul2003@cluster0.aibpfux.mongodb.net/proyecto_bigdata.viajes_chile_denomades") \
    .config("spark.mongodb.write.connection.uri", "mongodb+srv://AngeloRojo:azul2003@cluster0.aibpfux.mongodb.net/proyecto_bigdata.viajes_chile_denomades") \
    .config("spark.jars.packages", "org.mongodb.spark:mongo-spark-connector_2.12:10.1.1") \
    .get_session()

def ejecutar_procesamiento_semana6():
    print("🚀 Iniciando motor de procesamiento Spark (Semana 6)...")

    try:
        # 2. CARGA DE DATOS DESDE MONGODB ATLAS
        # Leemos la colección donde guardaste los registros de Denomades
        df = spark.read.format("mongodb").load()
        
        total_inicial = df.count()
        print(f"📊 Registros totales encontrados en Atlas: {total_inicial}")

        # 3. PROCESO DE LIMPIEZA (ETL)
        # - trim: Quita espacios en blanco accidentales.
        # - upper: Estandariza las ciudades (ej: 'santiago' y 'Santiago' pasan a ser 'SANTIAGO').
        # - dropDuplicates: Elimina tours repetidos para asegurar datos únicos.
        df_limpio = df.withColumn("actividad", trim(col("actividad"))) \
                      .withColumn("ubicacion", upper(trim(col("ubicacion")))) \
                      .dropDuplicates(["actividad"])

        # 4. FORMATEO DE PRECIOS
        # Convertimos el precio a tipo 'double' (decimal) para realizar cálculos matemáticos.
        # Filtramos precios mayores a 0 para evitar datos erróneos.
        df_final = df_limpio.withColumn("precio_num", col("precio_num").cast("double")) \
                            .filter(col("precio_num") > 0)

        total_final = df_final.count()
        print(f"✅ Limpieza completada con éxito.")
        print(f"♻️ Se procesaron {total_final} registros únicos.")
        print(f"🗑️ Se eliminaron {total_inicial - total_final} registros inconsistentes o duplicados.")

        # 5. ANÁLISIS DESCRIPTIVO (PREVIA SEMANA 7)
        # Agrupamos por ubicación y calculamos el precio promedio redondeado.
        print("\n💰 Precio Promedio de Tours por Ubicación (Ranking):")
        df_final.groupBy("ubicacion") \
                .agg(round(mean("precio_num"), 0).alias("Promedio_CLP")) \
                .sort(col("Promedio_CLP").desc()) \
                .show(15)

        # 6. VISUALIZACIÓN DE DATOS ESTANDARIZADOS
        print("📝 Muestra de los primeros 10 registros limpios:")
        df_final.select("actividad", "ubicacion", "precio_num").show(10, truncate=False)

    except Exception as e:
        print(f"❌ Error durante el procesamiento: {e}")
        print("💡 Consejo: Asegúrate de tener conexión a internet y que el Kernel de Jupyter esté activo.")

# --- INICIO DEL PROGRAMA ---
if __name__ == "__main__":
    ejecutar_procesamiento_semana6()
```

```python
# --- PROYECTO BIG DATA: SEMANA 6 ---
# Estudiante: Angelo Rojo
# Objetivo: Procesamiento y limpieza de datos (Plan B: PyMongo + Pandas)

import os
import sys
import pandas as pd
from pymongo import MongoClient
import certifi
from datetime import datetime

# 1. CONFIGURACIÓN DE CONEXIÓN A MONGODB ATLAS
uri = "mongodb+srv://AngeloRojo:azul2003@cluster0.aibpfux.mongodb.net/?appName=Cluster0"

def ejecutar_procesamiento_semana6():
    print("🚀 Iniciando motor de procesamiento (Semana 6 - Optimizado)...")
    
    try:
        # 2. CARGA DE DATOS DESDE MONGODB ATLAS
        client = MongoClient(uri, tlsCAFile=certifi.where())
        db = client['proyecto_bigdata']
        coleccion_origen = db['viajes_chile_denomades']
        coleccion_destino = db['viajes_chile_limpios'] # Nueva colección para datos pulidos
        
        print("🔍 Extrayendo datos de la nube...")
        datos_raw = list(coleccion_origen.find())
        
        if not datos_raw:
            print("⚠️ No se encontraron datos en Atlas. Verifica tu colección.")
            return

        # 3. CONVERSIÓN A DATAFRAME
        df = pd.DataFrame(datos_raw)
        total_inicial = len(df)
        print(f"📊 Registros totales encontrados: {total_inicial}")

        # 4. PROCESO DE LIMPIEZA (ETL)
        # Limpiamos espacios y estandarizamos ciudades
        df['actividad'] = df['actividad'].fillna('Actividad Desconocida').astype(str).str.strip()
        df['ubicacion'] = df['ubicacion'].fillna('CHILE').astype(str).str.strip().str.upper()
        
        # Ajustamos el filtro de duplicados: 
        # Si la actividad tiene el mismo nombre Y precio, se considera duplicado real.
        df_limpio = df.drop_duplicates(subset=['actividad', 'precio_num']).copy()

        # 5. FORMATEO Y FILTRADO DE PRECIOS
        df_limpio['precio_num'] = pd.to_numeric(df_limpio['precio_num'], errors='coerce')
        # Nos quedamos solo con los que tienen precios válidos y mayores a 0
        df_final = df_limpio[df_limpio['precio_num'] > 1000].copy() # Filtro de mínimo 1000 pesos

        total_final = len(df_final)
        print(f"✅ Limpieza completada con éxito.")
        print(f"♻️ Registros útiles para análisis: {total_final}")
        print(f"🗑️ Se descartaron {total_inicial - total_final} registros (duplicados o errores).")

        # 6. ANÁLISIS DESCRIPTIVO (Ranking de precios)
        print("\n💰 PRECIO PROMEDIO POR UBICACIÓN (RESUMEN):")
        ranking = df_final.groupby('ubicacion')['precio_num'].mean().round(0).sort_values(ascending=False)
        print(ranking)

        # 7. GUARDADO DE DATOS LIMPIOS (MANDATORIO PARA SEMANA 7)
        print(f"\n💾 Guardando {total_final} registros limpios en la colección 'viajes_chile_limpios'...")
        
        # Primero limpiamos la colección de destino para no duplicar si ejecutas el script dos veces
        coleccion_destino.delete_many({}) 
        
        # Convertimos el DataFrame a una lista de diccionarios para MongoDB
        # Quitamos la columna '_id' original de los datos raw para que Atlas genere IDs nuevos y limpios
        df_para_subir = df_final.drop(columns=['_id']) if '_id' in df_final.columns else df_final
        datos_subir = df_para_subir.to_dict('records')
        
        if datos_subir:
            coleccion_destino.insert_many(datos_subir)
            print("🎉 ¡Copia de seguridad limpia creada exitosamente en Atlas!")

        print("\n📝 MUESTRA DE DATOS FINALES:")
        print(df_final[['actividad', 'ubicacion', 'precio_num']].head(10).to_string(index=False))

    except Exception as e:
        print(f"❌ Error durante el procesamiento: {e}")
    finally:
        client.close()

# --- INICIO DEL PROGRAMA ---
if __name__ == "__main__":
    ejecutar_procesamiento_semana6()
```

```python
# --- PROYECTO BIG DATA: SEMANA 6 ---
# Estudiante: Angelo Rojo
# Objetivo: Procesamiento, limpieza y estandarización de datos (ETL)
# Resultado: Creación de la colección 'viajes_chile_limpios'

import pandas as pd
from pymongo import MongoClient
import certifi
from datetime import datetime

# 1. CONFIGURACIÓN DE CONEXIÓN A MONGODB ATLAS
uri = "mongodb+srv://AngeloRojo:azul2003@cluster0.aibpfux.mongodb.net/?appName=Cluster0"

def ejecutar_semana_6():
    print("🚀 INICIANDO PROCESAMIENTO SEMANA 6: LIMPIEZA DE DATOS")
    
    try:
        # 2. CONEXIÓN Y EXTRACCIÓN
        client = MongoClient(uri, tlsCAFile=certifi.where())
        db = client['proyecto_bigdata']
        
        # Leemos de la colección de captura (Semana 5)
        coleccion_raw = db['viajes_chile_denomades']
        # Guardaremos en la colección de limpieza (Semana 6)
        coleccion_clean = db['viajes_chile_limpios']
        
        print("🔍 Extrayendo registros de la Semana 5...")
        datos = list(coleccion_raw.find())
        df = pd.DataFrame(datos)
        
        if df.empty:
            print("⚠️ No hay datos para procesar.")
            return

        print(f"📊 Registros brutos encontrados: {len(df)}")

        # 3. TRANSFORMACIONES (LIMPIEZA)
        # Limpieza de textos y estandarización a mayúsculas
        df['actividad'] = df['actividad'].astype(str).str.strip()
        df['ubicacion'] = df['ubicacion'].astype(str).str.upper().str.strip()
        
        # Conversión de precios a formato numérico
        df['precio_num'] = pd.to_numeric(df['precio_num'], errors='coerce')

        # 4. ELIMINACIÓN DE DUPLICADOS (REGLA DE NEGOCIO)
        # Eliminamos registros donde se repita la actividad y el precio en la misma zona
        df_limpio = df.drop_duplicates(subset=['actividad', 'ubicacion', 'precio_num']).copy()
        
        # Filtramos solo registros con precios reales (mayores a $500)
        df_final = df_limpio[df_limpio['precio_num'] > 500].copy()

        print(f"✅ Proceso ETL finalizado.")
        print(f"♻️ Registros limpios generados: {len(df_final)}")

        # 5. CARGA DE DATOS LIMPIOS A ATLAS
        print(f"💾 Actualizando colección 'viajes_chile_limpios'...")
        
        # Limpiamos la colección de destino para evitar duplicados de ejecuciones previas
        coleccion_clean.delete_many({}) 
        
        # Preparamos para subir (quitando IDs internos de Mongo)
        if '_id' in df_final.columns:
            df_final = df_final.drop(columns=['_id'])
            
        datos_a_subir = df_final.to_dict('records')
        
        if datos_a_subir:
            coleccion_clean.insert_many(datos_subir)
            print("🎉 ¡Datos de la Semana 6 guardados exitosamente!")

        # 6. RESUMEN PARA EL REPORTE
        print("\n📈 RESUMEN DE PRECIOS PROMEDIO POR ZONA:")
        resumen = df_final.groupby('ubicacion')['precio_num'].mean().round(0)
        print(resumen)

    except Exception as e:
        print(f"❌ Error: {e}")
    finally:
        client.close()

if __name__ == "__main__":
    ejecutar_semana_6()
```
