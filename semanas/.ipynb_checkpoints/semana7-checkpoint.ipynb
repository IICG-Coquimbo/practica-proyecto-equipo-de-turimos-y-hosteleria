```python
# --- PROYECTO BIG DATA: SEMANA 7 ---
# Estudiante: Angelo Rojo
# Objetivo: Análisis Exploratorio (EDA) de datos limpios

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pymongo import MongoClient
import certifi

# 1. CONEXIÓN A LOS DATOS DE LA SEMANA 6
uri = "mongodb+srv://AngeloRojo:azul2003@cluster0.aibpfux.mongodb.net/?appName=Cluster0"

def generar_analisis_visual():
    print("📊 Generando Reporte Visual (Semana 7)...")
    
    try:
        client = MongoClient(uri, tlsCAFile=certifi.where())
        db = client['proyecto_bigdata']
        # Usamos la colección con los 13 registros únicos y validados
        df = pd.DataFrame(list(db['viajes_chile_limpios'].find()))
        
        if df.empty:
            print("⚠️ No hay datos limpios. Ejecuta el script de la Semana 6 primero.")
            return

        # 2. PREPARACIÓN ESTÉTICA
        sns.set_theme(style="whitegrid", palette="muted")
        fig, axes = plt.subplots(1, 2, figsize=(16, 7))

        # 3. GRÁFICO DE BARRAS: COMPARATIVA DE PRECIOS
        # Mostramos cuánto cuesta en promedio viajar a cada zona
        resumen = df.groupby('ubicacion')['precio_num'].mean().sort_values(ascending=False)
        sns.barplot(x=resumen.index, y=resumen.values, ax=axes[0], hue=resumen.index, legend=False)
        axes[0].set_title('Inversión Promedio por Destino ($ CLP)', fontsize=15, pad=20)
        axes[0].set_ylabel('Precio Promedio')
        axes[0].set_xlabel('Zona Geográfica')

        # 4. GRÁFICO DE TORTA: PARTICIPACIÓN DE MERCADO
        # Vemos qué porcentaje de la oferta total corresponde a cada zona
        distribucion = df['ubicacion'].value_counts()
        axes[1].pie(distribucion, labels=distribucion.index, autopct='%1.1f%%', 
                    startangle=90, colors=['#ff9999','#66b3ff'])
        axes[1].set_title('Distribución de la Oferta Turística', fontsize=15, pad=20)

        plt.tight_layout()
        print("🎨 Gráficos generados exitosamente.")
        plt.show()

        # 5. CONCLUSIÓN AUTOMATIZADA PARA EL INFORME
        p_san_pedro = df[df['ubicacion'] == 'SAN PEDRO DE ATACAMA']['precio_num'].mean()
        p_torres = df[df['ubicacion'] == 'TORRES DEL PAINE']['precio_num'].mean()
        
        print("\n📝 HALLAZGOS PARA EL REPORTE FINAL:")
        print(f"1. Se detectó una alta redundancia en la captura (aprox 97% de duplicados).")
        print(f"2. Torres del Paine es {p_torres/p_san_pedro:.1f} veces más caro que San Pedro de Atacama.")
        print(f"3. La base de datos 'viajes_chile_limpios' ya está lista para consumo grupal.")

    except Exception as e:
        print(f"❌ Error en Visualización: {e}")
    finally:
        client.close()

if __name__ == "__main__":
    generar_analisis_visual()
```
