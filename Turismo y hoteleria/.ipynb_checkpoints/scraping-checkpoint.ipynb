```python
from datetime import datetime
from pymongo import MongoClient
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# ==========================================================
# CONEXION MONGODB
# ==========================================================
try:
    client = MongoClient("mongodb://bigdata_mongodb:27017/")
    db = client["proyecto_bigdata"]
    coleccion = db["alojamientos"]
    print("Conexion a MongoDB exitosa!")
except Exception as e:
    print(f"Error conectando a MongoDB: {e}")

# ==========================================================
# CIUDADES A SCRAPEAR (CHILE)
# ==========================================================
ciudades = [
    "Arica",
    "Iquique",
    "Calama",
    "Antofagasta",
    "Copiapo",
    "La Serena",
    "Valparaiso",
    "Vina del Mar",
    "Santiago",
    "Rancagua",
    "Talca",
    "Chillan",
    "Concepcion",
    "Temuco",
    "Valdivia",
    "Puerto Varas",
    "Puerto Montt",
    "Coyhaique",
    "Puerto Natales",
    "Punta Arenas"
]

# ==========================================================
# FUNCIONES AUXILIARES
# ==========================================================
def limpiar_precio(texto):
    if not texto:
        return 0.0
    numeros = ""
    for c in texto:
        if c.isdigit() or c == '.':
            numeros += c
    try:
        return float(numeros) if numeros else 0.0
    except:
        return 0.0

def configurar_driver():
    opciones = Options()
    opciones.add_argument("--headless")
    opciones.add_argument("--no-sandbox")
    opciones.add_argument("--disable-dev-shm-usage")
    opciones.add_argument("--disable-blink-features=AutomationControlled")
    opciones.add_argument("--disable-gpu")
    opciones.add_argument("--window-size=1920,1080")

    opciones.add_argument(
        "user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120 Safari/537.36"
    )

    driver = webdriver.Chrome(
        service=Service(ChromeDriverManager().install()),
        options=opciones
    )
    return driver

def obtener_zona(ciudad):
    norte = [
        "Arica", "Iquique", "Calama",
        "Antofagasta", "Copiapo", "La Serena"
    ]
    centro = [
        "Valparaiso", "Vina del Mar",
        "Santiago", "Rancagua", "Talca"
    ]
    if ciudad in norte:
        return "Norte"
    elif ciudad in centro:
        return "Centro"
    else:
        return "Sur"

# ==========================================================
# SCRAPER TRIPADVISOR
# ==========================================================
def scraper_tripadvisor(ciudad):
    # URL para hoteles en Tripadvisor Chile
    url = f"https://www.tripadvisor.cl/Hotels-g294291-{ciudad.replace(' ', '_')}-Chile-Hotels.html"
    
    print("=" * 50)
    print("Ciudad:", ciudad)
    print("URL:", url)
    print("=" * 50)

    driver = None
    wait = None

    try:
        driver = configurar_driver()
        wait = WebDriverWait(driver, 15)
        driver.get(url)

        # Esperar carga inicial
        time.sleep(8)

        # Scroll para cargar más resultados
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(4)
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(4)

        # Buscar alojamientos (hoteles)
        alojamientos = driver.find_elements(
            By.CSS_SELECTOR,
            'div[data-testid="property-card"]'
        )

        # Si no encuentra con ese selector, probar alternativos
        if not alojamientos:
            alojamientos = driver.find_elements(
                By.CSS_SELECTOR,
                '.WAllg._T'
            )
        
        if not alojamientos:
            alojamientos = driver.find_elements(
                By.CSS_SELECTOR,
                'div.prw_rup.prw_metadata_click_tracker'
            )

        print("Resultados encontrados:", len(alojamientos))

        guardados = 0

        for i, item in enumerate(alojamientos[:20]):  # Limitar a 20 por ciudad
            try:
                # Intentar obtener nombre
                nombre_selectors = [
                    '.fIrG9._Q',
                    'a[data-testid="property-card-title"]',
                    '.HPQci .fIrG9',
                    'h3 a',
                    '.property_title'
                ]
                
                nombre = ""
                for selector in nombre_selectors:
                    try:
                        nombre_elem = item.find_element(By.CSS_SELECTOR, selector)
                        nombre = nombre_elem.text.strip()
                        if nombre:
                            break
                    except:
                        continue

                if not nombre or len(nombre) < 3:
                    continue

                # Buscar precio
                precio_selectors = [
                    '[data-testid="price-and-discounted-price"]',
                    '.price',
                    '.b',
                    '.offer-price',
                    '.ui_price'
                ]
                
                precio_texto = ""
                for selector in precio_selectors:
                    try:
                        precio_elem = item.find_element(By.CSS_SELECTOR, selector)
                        precio_texto = precio_elem.text.strip()
                        if precio_texto:
                            break
                    except:
                        continue

                precio = limpiar_precio(precio_texto)

                # Datos adicionales opcionales
                calificacion = 0.0
                try:
                    calif_elem = item.find_element(By.CSS_SELECTOR, 
                        '[data-testid="review-score"] span, .ui_bubble_rating'
                    )
                    calif_text = calif_elem.get_attribute('alt') or calif_elem.text
                    if 'bubble' in calif_text.lower():
                        calif_num = calif_text.split('_')[-1]
                        calificacion = float(calif_num) / 10 if calif_num.isdigit() else 0.0
                except:
                    pass

                registro = {
                    "nombre_hotel": nombre,
                    "precio_noche": precio,
                    "ciudad": ciudad,
                    "zona_geografica": obtener_zona(ciudad),
                    "plataforma": "tripadvisor.cl",
                    "calificacion": calificacion,
                    "fecha_captura": datetime.now(),
                    "integrante": "Juan_Salas",
                    "grupo": "G5_Turismo_Hoteleria"
                }

                coleccion.update_one(
                    {
                        "nombre_hotel": nombre,
                        "ciudad": ciudad,
                        "plataforma": "tripadvisor.cl"
                    },
                    {"$set": registro},
                    upsert=True
                )

                guardados += 1
                print(f"[{i+1}] Guardado -> {nombre[:45]} | ${precio} | {calificacion}⭐")

            except Exception as e:
                continue

        print(f"Total guardados en {ciudad}: {guardados}")
        return guardados

    except Exception as e:
        print("Error en scraper:", e)
        return 0

    finally:
        if driver:
            driver.quit()

# ==========================================================
# EJECUCION GENERAL
# ==========================================================
total_antes = coleccion.count_documents({"plataforma": "tripadvisor.cl"})
print("Registros Tripadvisor antes:", total_antes)

total_nuevos = 0

for i, ciudad in enumerate(ciudades):
    try:
        nuevos = scraper_tripadvisor(ciudad)
        total_nuevos += nuevos

        print("Esperando 15 segundos...\n")
        time.sleep(15)  # Más tiempo por ser Tripadvisor
        
        # Rotar ciudades para evitar bloqueos
        if (i + 1) % 5 == 0:
            print("Pausa extra de 60 segundos...\n")
            time.sleep(60)
            
    except KeyboardInterrupt:
        print("\nInterrumpido por usuario")
        break
    except Exception as e:
        print(f"Error procesando {ciudad}: {e}")
        continue

total_despues = coleccion.count_documents({"plataforma": "tripadvisor.cl"})

print("=" * 50)
print("SCRAPING TRIPADVISOR COMPLETADO")
print("=" * 50)
print("Antes:", total_antes)
print("Ahora:", total_despues)
print("Nuevos:", total_despues - total_antes)
print("=" * 50)
```
