```python
from datetime import datetime, timedelta
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time
import re
import json
import os
from pymongo import MongoClient
from pymongo.errors import ConnectionFailure, ServerSelectionTimeoutError

# ==========================================
# CONEXION MONGODB CORREGIDA
# ==========================================
def conectar_mongodb():
    try:
        client = MongoClient(
            "mongodb://bigdata_mongodb3:27017/",
            serverSelectionTimeoutMS=3000,
            connectTimeoutMS=3000
        )
        client.admin.command('ping')
        db = client["proyecto_bigdata"]
        coleccion = db["alojamientos"]
        print("✅ MongoDB conectado!")
        return coleccion, client, True
    except (ConnectionFailure, ServerSelectionTimeoutError):
        print("⚠️  MongoDB no disponible - Usando JSON")
        return None, None, False

coleccion, client, mongo_activo = conectar_mongodb()

# ==========================================
# CIUDADES EXPEDIA (nombres optimizados para Expedia)
# ==========================================
ciudades_expedia = {
    "Arica": "Arica-Chile", "Iquique": "Iquique-Chile", "Calama": "Calama-Chile", "Antofagasta": "Antofagasta-Chile", 
    "Copiapo": "Copiapo-Chile", "La Serena": "La-Serena-Chile", "Valparaiso": "Valparaiso-Chile", 
    "Vina del Mar": "Vina-del-Mar-Chile", "Santiago": "Santiago-Chile", "Rancagua": "Rancagua-Chile",
    "Talca": "Talca-Chile", "Chillan": "Chillan-Chile", "Concepcion": "Concepcion-Chile", "Temuco": "Temuco-Chile",
    "Valdivia": "Valdivia-Chile", "Puerto Varas": "Puerto-Varas-Chile", "Puerto Montt": "Puerto-Montt-Chile",
    "Coyhaique": "Coyhaique-Chile", "Puerto Natales": "Puerto-Natales-Chile", "Punta Arenas": "Punta-Arenas-Chile"
}

ciudades = list(ciudades_expedia.keys())

# ==========================================
# FUNCIONES JSON
# ==========================================
def guardar_json(registro):
    archivo = "expedia_hoteles.json"
    try:
        datos = []
        if os.path.exists(archivo):
            with open(archivo, 'r', encoding='utf-8') as f:
                datos = json.load(f)
        
        duplicado = any(
            r.get("nombre") == registro["nombre"] and 
            r.get("ciudad") == registro["ciudad"]
            for r in datos
        )
        
        if not duplicado:
            datos.append(registro)
            with open(archivo, 'w', encoding='utf-8') as f:
                json.dump(datos, f, ensure_ascii=False, indent=1)
            return True
        return False
    except Exception as e:
        print(f"❌ Error JSON: {e}")
        return False

def contar_json():
    try:
        if os.path.exists("expedia_hoteles.json"):
            with open("expedia_hoteles.json", 'r', encoding='utf-8') as f:
                datos = json.load(f)
                return len([r for r in datos if r.get("plataforma") == "expedia.com"])
        return 0
    except:
        return 0

# ==========================================
# CONTAR REGISTROS CORREGIDO
# ==========================================
def contar_registros():
    if mongo_activo and coleccion is not None:
        try:
            return coleccion.count_documents({"plataforma": "expedia.com"})
        except:
            print("⚠️  Error contando MongoDB, usando JSON")
    
    return contar_json()

# ==========================================
# CONFIG DRIVER - CORREGIDO ✅
# ==========================================
def configurar_driver():
    opciones = Options()
    opciones.add_argument("--headless")
    opciones.add_argument("--no-sandbox")
    opciones.add_argument("--disable-dev-shm-usage")
    opciones.add_argument("--disable-blink-features=AutomationControlled")
    opciones.add_argument("--window-size=1920,1080")
    opciones.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
    
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=opciones)
    
    # ✅ SCRIPT CORREGIDO - Sintaxis JavaScript válida
    driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")
    
    return driver

# ==========================================
# FUNCIONES AUXILIARES
# ==========================================
def extraer_precio_expedia(texto):
    if not texto:
        return 0
    # Expedia usa formatos como "$123.456", "123.456 CLP", etc.
    match = re.search(r'[\$]?([\d\.]+(?:,\d{2})?)', texto.replace(' ', ''))
    if match:
        precio_str = match.group(1).replace('.', '').replace(',', '.')
        try:
            return int(float(precio_str))
        except:
            pass
    return 0

def obtener_zona(ciudad):
    norte = ["Arica","Iquique","Calama","Antofagasta","Copiapo"]
    centro = ["La Serena","Valparaiso","Vina del Mar","Santiago","Rancagua"]
    sur = ["Talca","Chillan","Concepcion","Temuco","Valdivia"]
    
    if ciudad in norte: return "Norte"
    elif ciudad in centro: return "Centro"
    elif ciudad in sur: return "Sur"
    return "Austral"

def obtener_tipo_habitacion_expedia(card):
    """Extrae el tipo de habitación de Expedia"""
    tipos = ["Doble", "Individual", "Suite", "Familiar", "Estándar"]
    try:
        tipo_elems = card.find_elements(By.CSS_SELECTOR, 
            ".room-type, .room-name, .room-title, [class*='room'], .offer-room"
        )
        for elem in tipo_elems:
            tipo = elem.text.strip().lower()
            if any(palabra in tipo for palabra in ["doble", "double", "twin", "2 cama"]):
                return "Doble"
            elif any(palabra in tipo for palabra in ["individual", "single", "1 cama"]):
                return "Individual"
            elif "suite" in tipo:
                return "Suite"
            elif any(palabra in tipo for palabra in ["familiar", "family"]):
                return "Familiar"
        return "Estándar"
    except:
        return "Doble"  # Default

# ==========================================
# SCRAPER PRINCIPAL EXPEDIA - ADAPTADO
# ==========================================
def scraper_expedia(ciudad):
    nombre_expedia = ciudades_expedia[ciudad]
    checkin = (datetime.now() + timedelta(days=7)).strftime("%Y-%m-%d")
    checkout = (datetime.now() + timedelta(days=10)).strftime("%Y-%m-%d")
    
    # URL Expedia optimizada para Chile
    url = f"https://www.expedia.cl/Hotel-Search?destination={nombre_expedia}&startDate={checkin}&endDate={checkout}&adults=2&sort=RECOMMENDED"

    print(f"\n🏨 [{ciudades.index(ciudad)+1:2d}/{len(ciudades)}] {ciudad} ({nombre_expedia})")
    print(f"📅 {checkin} → {checkout}")

    driver = configurar_driver()
    
    guardados = 0
    try:
        driver.get(url)
        print("   ⏳ Esperando carga...")
        
        # Esperar carga de propiedades Expedia
        WebDriverWait(driver, 20).until(
            lambda d: len(d.find_elements(By.CSS_SELECTOR, 
                "[data-testid='property-card'], .uitk-card, .property-card, [data-stid='property-list-item']")) > 0
        )
        time.sleep(5)

        # Scroll para cargar más propiedades
        for _ in range(6):
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(2)

        # Selectores específicos de Expedia
        cards = driver.find_elements(By.CSS_SELECTOR, 
            "[data-testid='property-card'], "
            ".uitk-card, "
            ".property-card, "
            "[data-stid='property-list-item'], "
            "div[class*='property-card'], "
            ".listing-card"
        )
        
        print(f"   📊 {len(cards)} cards encontradas")

        for i, card in enumerate(cards[:20]):  # 20 máximo por ciudad
            try:
                # NOMBRE DEL HOTEL - Selectores Expedia
                nombre = "Hotel"
                nombre_selectors = [
                    "[data-testid='title']",
                    ".uitk-heading",
                    ".property-name",
                    ".property-title",
                    "h3",
                    "[data-stid='property-list-item-title']"
                ]
                for selector in nombre_selectors:
                    try:
                        nombre_elem = card.find_element(By.CSS_SELECTOR, selector)
                        nombre = nombre_elem.text.strip()
                        if nombre and len(nombre) > 3:
                            break
                    except:
                        continue
                nombre = nombre[:70] or "Hotel"

                # PRECIO - Selectores Expedia
                precio = 0
                precio_selectors = [
                    "[data-testid='price-and-discounted-price']",
                    ".price",
                    ".total-price",
                    "[class*='price']",
                    ".lead-price",
                    "[data-stid='price']"
                ]
                for selector in precio_selectors:
                    try:
                        precio_elem = card.find_element(By.CSS_SELECTOR, selector)
                        precio = extraer_precio_expedia(precio_elem.text)
                        if precio > 0:
                            break
                    except:
                        continue

                # RATING / PUNTUACION - Expedia
                puntuacion = 0
                rating_selectors = [
                    "[data-testid='reviews']",
                    ".guest-rating",
                    ".review-score",
                    "[class*='review']",
                    "[class*='rating']"
                ]
                for selector in rating_selectors:
                    try:
                        rating_elem = card.find_element(By.CSS_SELECTOR, selector)
                        match = re.search(r'([\d.]+)', rating_elem.text)
                        if match:
                            puntuacion = float(match.group())
                            break
                    except:
                        continue

                # ESTRELLAS
                estrellas = 0
                if puntuacion >= 9.0: estrellas = 5
                elif puntuacion >= 8.0: estrellas = 4
                elif puntuacion >= 7.0: estrellas = 3
                elif puntuacion > 0: estrellas = 2

                # TIPO HABITACION
                tipo_habitacion = obtener_tipo_habitacion_expedia(card)

                # REGISTRO
                registro = {
                    "integrante": "Juan-Salas",
                    "grupo": "G5_Turismo_Hoteleria",
                    "nombre": nombre,
                    "ciudad": ciudad,
                    "codigo_iata": nombre_expedia.replace("-", "").replace("Chile", "").upper()[:3],
                    "zona_geografica": obtener_zona(ciudad),
                    "precio": precio,
                    "precio_clp": precio,
                    "estrellas": estrellas,
                    "puntuacion": round(puntuacion, 1),
                    "adultos": 2,
                    "noches": 3,
                    "tipo_habitacion": tipo_habitacion,
                    "checkin": checkin,
                    "checkout": checkout,
                    "url_origen": url,
                    "plataforma": "expedia.com",
                    "fecha": datetime.now().isoformat()
                }

                # GUARDAR
                guardado = False
                if mongo_activo and coleccion is not None:
                    try:
                        coleccion.insert_one(registro)
                        guardado = True
                    except:
                        pass

                if not guardado:
                    guardado = guardar_json(registro)

                if guardado and precio > 0:
                    guardados += 1
                    print(f"   ✅ {i+1:2d}. {nombre[:35]}... ${precio:,} ⭐{puntuacion:.1f} ({tipo_habitacion})")

            except Exception as e:
                continue

    except Exception as e:
        print(f"   ❌ Error general: {str(e)[:80]}")
    
    finally:
        try:
            driver.quit()
        except:
            pass
    
    return guardados

# ==========================================
# MAIN
# ==========================================
print("🚀 EXPEDIA HOTELS SCRAPER v3.2 ✅")
print(f"🗄️  MongoDB: {'✅' if mongo_activo else '❌ (JSON)'}")

antes = contar_registros()
print(f"📈 Antes: {antes:,}")

total = 0
for i, ciudad in enumerate(ciudades):
    guardados = scraper_expedia(ciudad)
    total += guardados
    print(f"💾 {ciudad}: +{guardados} (Total: {total})")
    
    if i < len(ciudades) - 1:
        time.sleep(12)

despues = contar_registros()

print("\n" + "═"*60)
print("🎉 FINALIZADO")
print("═"*60)
print(f"📊 Antes:  {antes:,}")
print(f"📊 Ahora:  {despues:,}")
print(f"📈 Total:  {total:,}")
print(f"💾 JSON: expedia_hoteles.json")
print("═"*60)
```
