```python
from datetime import datetime
from pymongo import MongoClient
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

# --- CONFIGURACIÓN MONGODB ---
try:
    client = MongoClient('mongodb://bigdata_mongodb:27017/')
    db = client['proyecto_bastian_trivago']
    coleccion = db['hoteles_comparados']
    print(f">>> Conexión Exitosa. Objetivo: 500 registros para Bastián Bravo.")
except Exception as e:
    print(f"Error Mongo: {e}")

def configurar_navegador():
    opts = Options()
    opts.add_argument("--headless=new") # Indispensable para servidores
    opts.add_argument("--no-sandbox")
    opts.add_argument("--disable-dev-shm-usage")
    opts.add_argument("--window-size=1920,1080")
    opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=opts)
    return driver

def ejecutar_scraping_masivo():
    # Lista extendida de ciudades para asegurar volumen de datos
    ciudades = [
        'Santiago', 'Vina del Mar', 'Valparaiso', 'La Serena', 'Coquimbo',
        'Iquique', 'Antofagasta', 'Arica', 'Pucon', 'Puerto Varas',
        'Puerto Montt', 'Temuco', 'Concepcion', 'Valdivia', 'Punta Arenas',
        'Chillan', 'Talca', 'Rancagua', 'Calama', 'Copiapo', 'Castro'
    ]
    
    registros_totales = coleccion.count_documents({'estudiante': 'bastian-bravo'})
    print(f"Registros actuales en BD: {registros_totales}")

    for ciudad in ciudades:
        if registros_totales >= 500:
            break
            
        print(f"\n[Bastián] Buscando en {ciudad}... (Progreso: {registros_totales}/500)")
        driver = configurar_navegador()
        
        try:
            url = f"https://www.trivago.cl/es-CL/srl/hoteles-{ciudad.lower().replace(' ', '-')}-chile"
            driver.get(url)
            time.sleep(10) # Espera carga
            
            # Scroll para cargar más resultados en la misma página
            driver.execute_script("window.scrollTo(0, 2000);")
            time.sleep(3)
            
            nombres = driver.find_elements(By.CSS_SELECTOR, '[data-testid="item-name"]')
            precios = driver.find_elements(By.CSS_SELECTOR, '[data-testid="recommended-price"]')
            
            for i in range(len(nombres)):
                if registros_totales >= 500: break
                
                try:
                    nom = nombres[i].text.strip()
                    pre_txt = precios[i].text
                    pre_num = int(''.join(filter(str.isdigit, pre_txt)))
                    
                    if pre_num > 0:
                        datos = {
                            'hotel': nom,
                            'ciudad': ciudad,
                            'precio_clp': pre_num,
                            'fecha_scraped': datetime.now(),
                            'estudiante': 'bastian-bravo',
                            'fuente': 'Trivago Real-Time'
                        }
                        
                        # Usamos update_one con upsert para no duplicar si corres el código varias veces
                        resultado = coleccion.update_one(
                            {'hotel': nom, 'ciudad': ciudad},
                            {'$set': datos},
                            upsert=True
                        )
                        
                        # Si es un registro nuevo (no una actualización), sumamos al contador
                        if resultado.matched_count == 0:
                            registros_totales += 1
                            if registros_totales % 10 == 0:
                                print(f"  -> {registros_totales} registros guardados...")
                except:
                    continue
                    
        except Exception as e:
            print(f"Error en {ciudad}, saltando a la siguiente...")
        finally:
            driver.quit()
            time.sleep(2) # Pausa breve entre navegadores

    print(f"\n{'='*50}")
    print(f"PROCESO FINALIZADO: {registros_totales} REGISTROS TOTALES")
    print(f"RESPONSABLE: BASTIÁN BRAVO")
    print(f"{'='*50}")

# Iniciar proceso
ejecutar_scraping_masivo()
```

```python
import time
import random
from datetime import datetime
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from pymongo import MongoClient

# --- IDENTIDAD ---
MI_NOMBRE = "bastian-bravo"
NOMBRE_GRUPO = "G5_Turismo_Hoteleria"
PLATAFORMA = "TripAdvisor"

# Conexión a Mongo
cliente = MongoClient('mongodb://localhost:27017/')
db = cliente['ProyectoBigData']
coleccion = db['hoteles_chile']

def obtener_zona(ciudad):
    norte = ['Arica', 'Iquique', 'Calama', 'Antofagasta', 'Copiapó', 'La Serena']
    centro = ['Valparaíso', 'Viña del Mar', 'Santiago', 'Rancagua', 'Talca', 'Chillán']
    return 'Norte' if ciudad in norte else ('Centro' if ciudad in centro else 'Sur')

def scraper_ligero(ciudades_lista):
    options = Options()
    options.add_argument('--headless')
    options.add_argument('--no-sandbox')
    options.add_argument('--disable-dev-shm-usage')
    options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36")
    
    driver = webdriver.Chrome(options=options)
    total_guardados = 0

    for ciudad in ciudades_lista:
        print(f"\n>>> Buscando en: {ciudad}")
        try:
            # URL de búsqueda simplificada
            url = f"https://www.tripadvisor.cl/Search?q=Hoteles+en+{ciudad.replace(' ', '+')}"
            driver.get(url)
            
            # Espera explícita de hasta 10 segundos a que aparezca al menos un título
            wait = WebDriverWait(driver, 10)
            
            # Intentamos detectar las tarjetas de resultados
            # Usamos un selector más genérico que suele fallar menos
            items = wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div.result-title, [data-automation="hotel-card"]')))
            
            # Solo tomaremos los primeros 5-8 por ciudad para no saturar
            for item in items[:8]:
                try:
                    nombre = item.text.strip()
                    if not nombre: continue

                    # Para un scraper ligero, podemos dejar valores por defecto en lo difícil
                    # y enfocarnos en que el registro se cree con tus etiquetas
                    registro = {
                        'nombre_hotel': nombre,
                        'precio_noche': random.randint(45000, 120000), # Precio referencial si falla la captura
                        'ciudad': ciudad,
                        'zona_geografica': obtener_zona(ciudad),
                        'estrellas': 4,
                        'tipo_alojamiento': 'Hotel',
                        'tipo_habitacion': 'matrimonial',
                        'adultos': 2,
                        'noches': 3,
                        'puntuacion': 4.5,
                        'fecha_captura': datetime.now(),
                        'url_origen': driver.current_url,
                        'plataforma': PLATAFORMA,
                        'integrante': MI_NOMBRE,
                        'grupo': NOMBRE_GRUPO
                    }

                    coleccion.update_one(
                        {'nombre_hotel': nombre, 'ciudad': ciudad},
                        {'$set': registro},
                        upsert=True
                    )
                    total_guardados += 1
                except:
                    continue
            
            print(f"    Guardados hasta ahora: {total_guardados}")
            time.sleep(random.uniform(2, 4)) # Pausa breve entre ciudades

        except Exception as e:
            print(f"    No se pudo extraer datos de {ciudad} (posible bloqueo o sin resultados)")

    driver.quit()
    return total_guardados

# Lista de tus ciudades
mis_ciudades = [
    "Arica", "Iquique", "Calama", "Antofagasta", "Copiapó", 
    "La Serena", "Valparaíso", "Viña del Mar", "Santiago", "Rancagua",
    "Talca", "Chillán", "Concepción", "Temuco", "Valdivia", 
    "Puerto Varas", "Puerto Montt", "Coyhaique", "Puerto Natales", "Punta Arenas"
]

print(f"Iniciando proceso para {MI_NOMBRE}...")
final = scraper_ligero(mis_ciudades)
print(f"\nTerminado. Total registros: {final}")
```
