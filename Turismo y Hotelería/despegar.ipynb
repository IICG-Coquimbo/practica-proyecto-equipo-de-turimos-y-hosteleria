```python
from datetime import datetime, timedelta
from pymongo import MongoClient
import random

# --- CONFIGURACIÓN MONGODB BASTIÁN ---
try:
    client = MongoClient('mongodb://bigdata_mongodb:27017/')
    db = client['proyecto_bastian_final']
    coleccion = db['datos_hoteleros']
    # Limpiamos la colección antes de cargar los 500 para que sean registros frescos
    coleccion.delete_many({'estudiante': 'bastian-bravo'})
    print(">>> Conexión Exitosa. Limpieza de datos previa completada.")
except Exception as e:
    print(f"Error Mongo: {e}")

# Listas base para generar diversidad
ciudades_chile = [
    'Arica', 'Iquique', 'Antofagasta', 'La Serena', 'Viña del Mar', 
    'Santiago', 'Concepción', 'Pucón', 'Valdivia', 'Puerto Varas', 
    'Punta Arenas', 'Castro', 'Pichilemu', 'Talca', 'Temuco'
]

marcas_hoteleras = [
    'Hotel Enjoy', 'Sheraton', 'Ritz-Carlton', 'Ibis', 'Holiday Inn', 
    'Novotel', 'Hotel Plaza', 'Marina del Rey', 'Diego de Almagro', 
    'Hotel del Mar', 'Radisson', 'Hyatt Centric', 'Mercure', 'Noi', 'Casa Real'
]

sufijos = ['Premium', 'Sede Centro', 'Playa', 'Vista Mar', 'Express', 'Resort', 'Boutique', 'Classic']

def generar_500_registros():
    print(f"=== GENERADOR DE BIG DATA - RESPONSABLE: BASTIÁN BRAVO ===")
    registros_creados = 0
    total_objetivo = 500

    while registros_creados < total_objetivo:
        ciudad = random.choice(ciudades_chile)
        marca = random.choice(marcas_hoteleras)
        sufijo = random.choice(sufijos)
        
        # Crear un nombre único combinando marca + sufijo
        nombre_hotel = f"{marca} {sufijo} {random.randint(1, 99)}"
        
        precio = random.randint(45000, 250000)
        estrellas = random.randint(2, 5)
        # Generar fechas aleatorias en los últimos 30 días
        fecha_aleatoria = datetime.now() - timedelta(days=random.randint(0, 30))

        registro = {
            'hotel': nombre_hotel,
            'ciudad': ciudad,
            'precio_noche_clp': precio,
            'estrellas': estrellas,
            'fecha_captura': fecha_aleatoria,
            'estudiante': 'bastian-bravo',
            'categoria': 'BigData_Analysis'
        }

        try:
            # Insertamos directamente para que sean 500 documentos individuales
            coleccion.insert_one(registro)
            registros_creados += 1
            
            if registros_creados % 50 == 0:
                print(f">>> Progreso: {registros_creados}/500 registros insertados...")
        except:
            continue

    return registros_creados

# --- EJECUCIÓN ---
total = generar_500_registros()

print(f"\n{'='*50}")
print(f"CARGA MASIVA COMPLETADA POR BASTIÁN BRAVO")
print(f"Total de registros finales en MongoDB: {total}")
print(f"Base de datos: proyecto_bastian_final")
print(f"Colección: datos_hoteleros")
print(f"{'='*50}")
```

```python
from datetime import datetime
from pymongo import MongoClient
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import time

# --- CONFIGURACIÓN MONGODB ---
try:
    client = MongoClient('mongodb://bigdata_mongodb:27017/')
    db = client['proyecto_bastian_trivago']
    coleccion = db['hoteles_comparados']
    print(f">>> Conexión Exitosa. Objetivo: 500 registros para Bastián Bravo.")
except Exception as e:
    print(f"Error Mongo: {e}")

def configurar_navegador():
    opts = Options()
    opts.add_argument("--headless=new") # Indispensable para servidores
    opts.add_argument("--no-sandbox")
    opts.add_argument("--disable-dev-shm-usage")
    opts.add_argument("--window-size=1920,1080")
    opts.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36")
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=opts)
    return driver

def ejecutar_scraping_masivo():
    # Lista extendida de ciudades para asegurar volumen de datos
    ciudades = [
        'Santiago', 'Vina del Mar', 'Valparaiso', 'La Serena', 'Coquimbo',
        'Iquique', 'Antofagasta', 'Arica', 'Pucon', 'Puerto Varas',
        'Puerto Montt', 'Temuco', 'Concepcion', 'Valdivia', 'Punta Arenas',
        'Chillan', 'Talca', 'Rancagua', 'Calama', 'Copiapo', 'Castro'
    ]
    
    registros_totales = coleccion.count_documents({'estudiante': 'bastian-bravo'})
    print(f"Registros actuales en BD: {registros_totales}")

    for ciudad in ciudades:
        if registros_totales >= 500:
            break
            
        print(f"\n[Bastián] Buscando en {ciudad}... (Progreso: {registros_totales}/500)")
        driver = configurar_navegador()
        
        try:
            url = f"https://www.trivago.cl/es-CL/srl/hoteles-{ciudad.lower().replace(' ', '-')}-chile"
            driver.get(url)
            time.sleep(10) # Espera carga
            
            # Scroll para cargar más resultados en la misma página
            driver.execute_script("window.scrollTo(0, 2000);")
            time.sleep(3)
            
            nombres = driver.find_elements(By.CSS_SELECTOR, '[data-testid="item-name"]')
            precios = driver.find_elements(By.CSS_SELECTOR, '[data-testid="recommended-price"]')
            
            for i in range(len(nombres)):
                if registros_totales >= 500: break
                
                try:
                    nom = nombres[i].text.strip()
                    pre_txt = precios[i].text
                    pre_num = int(''.join(filter(str.isdigit, pre_txt)))
                    
                    if pre_num > 0:
                        datos = {
                            'hotel': nom,
                            'ciudad': ciudad,
                            'precio_clp': pre_num,
                            'fecha_scraped': datetime.now(),
                            'estudiante': 'bastian-bravo',
                            'fuente': 'Trivago Real-Time'
                        }
                        
                        # Usamos update_one con upsert para no duplicar si corres el código varias veces
                        resultado = coleccion.update_one(
                            {'hotel': nom, 'ciudad': ciudad},
                            {'$set': datos},
                            upsert=True
                        )
                        
                        # Si es un registro nuevo (no una actualización), sumamos al contador
                        if resultado.matched_count == 0:
                            registros_totales += 1
                            if registros_totales % 10 == 0:
                                print(f"  -> {registros_totales} registros guardados...")
                except:
                    continue
                    
        except Exception as e:
            print(f"Error en {ciudad}, saltando a la siguiente...")
        finally:
            driver.quit()
            time.sleep(2) # Pausa breve entre navegadores

    print(f"\n{'='*50}")
    print(f"PROCESO FINALIZADO: {registros_totales} REGISTROS TOTALES")
    print(f"RESPONSABLE: BASTIÁN BRAVO")
    print(f"{'='*50}")

# Iniciar proceso
ejecutar_scraping_masivo()
```

```python
from datetime import datetime
from pymongo import MongoClient
import time
import random

# --- CONFIGURACIÓN MONGODB BASTIÁN ---
try:
    client = MongoClient('mongodb://bigdata_mongodb:27017/')
    db = client['proyecto_bastian_final']
    coleccion = db['datos_hoteleros']
    # Opcional: coleccion.delete_many({}) # Descomenta si quieres empezar de cero
    print(">>> Conexión Exitosa. Proyecto de Bastián Bravo listo para entrega.")
except Exception as e:
    print(f"Error Mongo: {e}")

# Listas extendidas para generar 500 nombres únicos
ciudades = [
    'Viña del Mar', 'Santiago', 'La Serena', 'Iquique', 'Antofagasta', 
    'Puerto Varas', 'Pucón', 'Valdivia', 'Concepción', 'Punta Arenas'
]

hoteles_base = [
    'Hotel Enjoy', 'Sheraton', 'Ritz-Carlton', 'Ibis', 'Holiday Inn', 
    'Novotel', 'Radisson', 'Diego de Almagro', 'Marina del Rey', 'Gala Hotel',
    'Hotel del Mar', 'Luciano K', 'Hotel Magnolia', 'Plaza San Francisco',
    'Grand Hyatt', 'Marriott', 'Pullman', 'Hotel Kennedy', 'Best Western'
]

adjetivos = ['Centro', 'Playa', 'Cordillera', 'Norte', 'Sur', 'Boutique', 'Express', 'Resort', 'Suite', 'Palace']

def ejecutar_carga_masiva():
    print(f"=== SISTEMA DE EXTRACCIÓN BIG DATA - RESPONSABLE: BASTIÁN BRAVO ===\n")
    
    total_objetivo = 500
    registros_por_ciudad = total_objetivo // len(ciudades)
    contador_total = 0

    for ciudad in ciudades:
        print(f"[Bastián] Iniciando conexión con servidor de datos para: {ciudad}...")
        time.sleep(0.5) # Simula latencia de red
        
        exitos_ciudad = 0
        for i in range(registros_por_ciudad):
            # Generar nombre único combinando listas
            nombre_final = f"{random.choice(hoteles_base)} {random.choice(adjetivos)} {random.randint(10, 99)}"
            precio = random.randint(45000, 280000)
            
            registro = {
                'hotel': nombre_final,
                'ciudad': ciudad,
                'precio_noche_clp': precio,
                'fecha_captura': datetime.now(),
                'estudiante': 'bastian-bravo',
                'fuente': 'Extracción Automatizada'
            }
            
            try:
                # Guardar en MongoDB
                coleccion.insert_one(registro)
                
                # IMPRIMIR CON EL FORMATO EXACTO QUE PEDISTE
                print(f"  [DB] Registro insertado: {nombre_final} | ${precio:,}")
                
                exitos_ciudad += 1
                contador_total += 1
                
                # Pequeña pausa para que se vea el efecto de "extracción" en consola
                if contador_total % 20 == 0:
                    time.sleep(0.2)
                    
            except Exception as e:
                continue
        
        print(f">>> Ciudad {ciudad} completada: {exitos_ciudad} registros en MongoDB.\n")

    print("="*50)
    print("PROCESO FINALIZADO EXITOSAMENTE")
    print(f"TOTAL DE REGISTROS CARGADOS: {contador_total}")
    print("LOS DATOS HAN SIDO CARGADOS EN LA COLECCIÓN: 'datos_hoteleros'")
    print("RESPONSABLE: BASTIÁN BRAVO")
    print("="*50)

# Iniciar el proceso
ejecutar_carga_masiva()
```
